Downloading and unzipping the files of year 2016 from the S3 bucket

```python
# Importing os, urllib, and zipfile libraries
import os
import urllib.request
import zipfile

# Specifying the year I want to download csv for
year = "2016"
month_names = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

# Specifying the S3 bucket url
s3_url = "https://s3.amazonaws.com/tripdata/"

# Create a directory to store the downloaded files
if not os.path.exists("BikeData"):
    os.makedirs("BikeData")

# Loop through the list of file names and download them to the data directory
for month in month_names:
    file_name = f"JC-{year}{month}-citibike-tripdata.csv.zip"
    url = s3_url + file_name
    save_path = os.path.join("BikeData", file_name)
    urllib.request.urlretrieve(url, save_path)
    print(f"{file_name} downloaded")

    # extract the contents of the zip file
    with zipfile.ZipFile(save_path, "r") as zip_ref:
        zip_ref.extractall("BikeData")
    print(f"{file_name} unzipped")

print("All files downloaded and unzipped successfully")
```

Downloading, unzipping, and concatenating files in one go

```python
# Importing os, urllib, and zipfile libraries
import os
import urllib.request
import zipfile

# Specifying the year I want to download csv for
year1 = "2018"
year2 = "2019"
month_names = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
dfs = []

# Specifying the S3 bucket url
s3_url = "https://s3.amazonaws.com/tripdata/"

# Create a directory to store the downloaded files
if not os.path.exists("BikeData1"):
    os.makedirs("BikeData1")

# Loop through the list of file names and download them to the data directory
for month in month_names:
    file_name = f"JC-{year1}{month}-citibike-tripdata.csv.zip"
    url = s3_url + file_name
    save_path = os.path.join("BikeData1", file_name)
    urllib.request.urlretrieve(url, save_path)
    print(f"{file_name} downloaded")

    # extract the contents of the zip file
    with zipfile.ZipFile(save_path, "r") as zip_ref:
        zip_ref.extractall("BikeData1")
    print(f"{file_name} unzipped")

    df = pd.read_csv(save_path)
    dfs.append(df)
 

for month in month_names:
    file_name = f"JC-{year2}{month}-citibike-tripdata.csv.zip"
    url = s3_url + file_name
    save_path = os.path.join("BikeData1", file_name)
    urllib.request.urlretrieve(url, save_path)
    print(f"{file_name} downloaded")

    # extract the contents of the zip file
    with zipfile.ZipFile(save_path, "r") as zip_ref:
        zip_ref.extractall("BikeData1")
    print(f"{file_name} unzipped")

    df = pd.read_csv(save_path)
    dfs.append(df)

print("All files downloaded and unzipped successfully")

# concatenate the dataframes into one
merged_df = pd.concat(dfs, ignore_index=True)

# write the merged dataframe to a new CSV file
merged_df.to_csv("merged_file.csv", index=False)

print("All csv files have been merged into one")
```

b. What is the average distance that bikers ride?

```python
from math import radians, sin, cos, sqrt, asin

# define a function to calculate distance using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6372.8 # Earth radius in kilometers
    
    # convert to radians
    d_lat = radians(lat2 - lat1)
    d_lon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    # apply Haversine formula
    a = sin(d_lat/2)**2 + cos(lat1)*cos(lat2)*sin(d_lon/2)**2
    c = 2*asin(sqrt(a))
    
    # calculate distance
    distance = R * c
    return distance
```

```python

# calculate distance between coordinates using haversine function
bd_1['distance_km'] = bd_1.apply(lambda row: haversine(row['start station latitude'], row['start station longitude'], row['end station latitude'], row['end station longitude']), axis=1)
bd_1
```

a. which is the most popular station?

```python
from pyspark.sql.functions import count
from pyspark.sql.functions import desc
from pyspark.sql.functions import col

# group the DataFrame by 'start station name' and count the number of occurrences of 'start station id'
grouped_df = df.groupBy('start station name').agg(count('start station id').alias('No_of_trips'))

# sort the DataFrame by 'No_of_trips' in descending order
grouped_df = grouped_df.sort(desc('No_of_trips'))

# select the first row from the DataFrame and extract the value of 'start station name'
most_pop_stn = grouped_df.select('start station name').first()[0]

print("Most popular station in 2018-2019 is {}".format(most_pop_stn))
```

b. What is the average distance that bikers ride?

```python
from pyspark.sql.functions import radians, sin, cos, sqrt, asin, col

R = 6372.8 # Earth radius in kilometers

# calculate distance in kilometers
df = df.withColumn("lat1", radians(col("start station latitude"))) \
       .withColumn("lon1", radians(col("start station longitude"))) \
       .withColumn("lat2", radians(col("end station latitude"))) \
       .withColumn("lon2", radians(col("end station longitude")))

df = df.withColumn("d_lat", col("lat2") - col("lat1")) \
       .withColumn("d_lon", col("lon2") - col("lon1")) \
       .withColumn("a", sin(col("d_lat")/2)**2 + cos(col("lat1"))*cos(col("lat2"))*sin(col("d_lon")/2)**2) \
       .withColumn("c", 2*asin(sqrt(col("a")))) \
       .withColumn("distance_km", R*col("c"))

# calculate average distance
avg_distance = df.selectExpr("avg(distance_km)").collect()[0][0]
print(avg_distance, "km")
```
